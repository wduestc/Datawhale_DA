```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('task2.csv', encoding='gb18030')
df.head()
```

```python
y = df['status']
X = df.drop('status', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, y, test_size=0.3, random_state=2018)
```

### 参数调优

```python
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
param = {'n_estimators':[10, 500], 'max_depth':[2, 3]}
clf = GridSearchCV(model, param_grid = param, cv=5)
clf.fit(Xtrain, Ytrain)
```

```python
clf.cv_results_
```

```python
# 获取对应的参数
print (clf.best_params_, clf.best_score_)
pd.DataFrame(clf.cv_results_).loc[:, ['params', 'mean_test_score', 'std_test_score']]
```
